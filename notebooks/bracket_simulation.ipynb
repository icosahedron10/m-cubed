# Bracket simulation
Run Monte Carlo bracket simulations from the example bracket.

```python
import pandas as pd
from march_madness_model.bracket import load_bracket_csv
from march_madness_model.probabilities import SeedProbabilityModel
from march_madness_model.simulator import TournamentSimulator

bracket = load_bracket_csv("data/example/example_bracket_64.csv")
matchups = pd.read_csv("data/example/seed_matchup_probs.csv")
model = SeedProbabilityModel.from_dataframes(matchup_df=matchups)
TournamentSimulator(bracket, model).simulate(1000, random_seed=42).head()
```
